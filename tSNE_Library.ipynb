```python
# importing 20NG and MNIST datasets
from sklearn.datasets import fetch_openml, fetch_20newsgroups
# importing library tSNE
from sklearn.manifold import TSNE
# plotting package
import matplotlib.pyplot as plt
```

```python
mnist = fetch_openml('mnist_784')
X_mnist = mnist.data.to_numpy()/255.0
```

```python
X_mnist = X_mnist[:20000]
y_mnist = mnist.target.astype('int')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

categories = ["alt.atheism", "sci.med", "sci.electronics", "comp.graphics", "talk.politics.guns", "sci.crypt"]
newsgroups = fetch_20newsgroups(subset='all', categories=categories, shuffle=True, random_state=42,
                                    remove=("headers", "footers", "quotes"))
X_20ng = newsgroups.data
y_20ng = newsgroups.target
# Converting Text data to vectors
vectorizer = TfidfVectorizer(max_features=1000)
vectors = vectorizer.fit_transform(X_20ng)
```

# Running and Plotting tSNE on MNIST and 20NG datasets with perplexity 5

```python
tsne_5 = TSNE(n_components=2, learning_rate='auto', init='random', perplexity=5, n_jobs=-1)
transformed_X_mnist = tsne_5.fit_transform(X_mnist)
transformed_X_mnist.shape
```

```python
import numpy as np
colors = plt.cm.rainbow(np.linspace(0, 1, 10))
marker_map = {0 : 'o', 1: 'x', 2: '+', 3: '^', 4: 's', 5: 'p', 6:'h', 7:'*', 8:'8', 9:'d'}
plotted_cluster = set()
for i in range(len(transformed_X_mnist)):
    # label = kmeans.labels_[i]
    if y_mnist[i] not in plotted_cluster:
        plt.scatter(transformed_X_mnist[i, 0], transformed_X_mnist[i, 1], c=colors[int(y_mnist[i]) % len(colors)], marker=marker_map[y_mnist[i]], label=f'Cluster {int(y_mnist[i]) % len(colors)}')
        plotted_cluster.add(y_mnist[i])
    else:
        plt.scatter(transformed_X_mnist[i, 0], transformed_X_mnist[i, 1], c=colors[int(y_mnist[i]) % len(colors)], marker=marker_map[y_mnist[i]])
plt.legend()
```
